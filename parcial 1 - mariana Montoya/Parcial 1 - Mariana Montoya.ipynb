```python
import numpy as np
import matplotlib.pyplot as plt

#hallar raices, 3


def f(x):
    return 3*x**5 + 5*x**4 - x**3

def df(x):
    return 15*x**4 + 20*x**3 - 3*x**2

#3.10

dominio = np.linspace(-2,2,100)
def hallar_raices(f,x1,df):
    it = 50
    error = 1 * 10**(-5)
    j = 0
    Xn_1 = x1
    

    grafica = f(dominio)
    while j < it or abs(f(Xn_1)) > error:
        Xn = Xn_1 - f(Xn_1)/df(Xn_1)
        j =  j + 1
        Xn_1 = Xn
        print(f"iteracion:{j}\nXn: {Xn}\nfuncion: {f(Xn_1)}\n")
        plt.plot(dominio,grafica)
    plt.plot(Xn,0,c = 'red', marker = '*')
    plt.plot(dominio,Xn_1,f(Xn_1),dominio)
    plt.grid()
    plt.show()

figura = hallar_raices(f,dominio,df)
```
