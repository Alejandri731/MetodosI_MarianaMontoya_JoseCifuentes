```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

```python
def f(x):
    return 3*x**(5)+5*x**(4)-x**(3)
```

```python
x=np.linspace(-2,2,50)
y=f(x)
```

```python
plt.plot(x,y)
plt.axhline(y=0, color="red")
plt.axvline(-0.002145811941152424, color="red")
```

```python
def derivada(f,x,h=x[1]-x[0]):
    return (f(x+h)-f(x-h))/(2*h)
```

```python
df=derivada(f,x)
```

```python
def newton_method(f,df,xn):
    
    error=1

    while error>1e-8:
        xn1= xn - f(xn)/df(f,xn)
        error=abs(f(xn1))
        
        xn=xn1
      
 
    return xn
    
```

```python
roots=[]
x=np.linspace(-2,2,20)
for i in x:

    h=newton_method(f,derivada,i)
    
    if h not in roots:
        roots.append(h)
new_roots=[]
for i in range(0,len(roots)-1):
    if (abs(roots[i+1])-abs(roots[i])) < 1e-13:
        new_roots.append(roots[i])
new_roots
        
```
