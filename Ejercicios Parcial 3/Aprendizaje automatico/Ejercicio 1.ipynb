```python
#Aprendizaje automatico 1

import numpy as np
import matplotlib.pyplot as plt
import sympy as sym

def f(x,y):
    return x**2 - y**2 + 2*x

a = 1
b = 1
p = np.array([a,b])

```

```python
def gradiente(x,y):
#    Dx = (f(x+h,y) - f(x-h,y))/(2*h)
#    Dy = (f(x,y+h) - f(x,y-h))/(2*h)
    Gradient = np.array([2*x+2,-2*y])
    return Gradient
```

```python
def polinomio_taylor(x,y):
    return f(1,1) + gradiente(1,1)[0]*(x-1) + gradiente(1,1)[1]*(y-1)
```

```python
polinomio_taylor(2,2)
```

```python
x = np.linspace(-10, 10, 11)
y = np.linspace(-10, 10, 11)
X, Y = np.meshgrid(x, y)
funcion = f(X,Y)
#Z = polinomio_taylor(X, Y)
```

```python
polinomio = polinomio_taylor(X,Y)
```

```python
fig = plt.figure(figsize=(10, 8))
ax = plt.axes(projection='3d')
 
ax.plot_surface(X, Y, funcion, cmap='cool', alpha=0.8)
ax.plot_surface(X, Y, polinomio, cmap='spring', alpha=0.8)
ax.scatter(1, 1, 2, c=0.5)
 
ax.set_title('f(x,y) = x²-y²+2x', fontsize=14)
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y', fontsize=12)
ax.set_zlabel('z', fontsize=12) 
plt.show()
```
